```python
df_3229 = pd.read_csv('/opt/ml/input/final_project/MBTI_merge_movieLens_3229_movie.csv')
side_info_df = pd.read_csv('/opt/ml/input/final_project/Crawling/imdb_side_df_EmptyList_to_None.csv')
```

```python
df_3229_plus_side_info = pd.merge(df_3229,side_info_df, how='left', on='Contents')
df_3229_plus_side_info
```

```python
translator = googletrans.Translator()

# def trans_to_ko(s):
#     out = translator.translate(s,dest='ko', src='en')
#     return out.text

def trans_to_ko(s):
    if s and isinstance(s, str):
        out = translator.translate(s,dest='ko', src='en')
        time.sleep(0.3)
        return out.text
    else:
        return None
```

```python
df_fillna_no_du = df_3229_plus_side_info.drop_duplicates(subset='movieId')
df_fillna_no_du = df_fillna_no_du.drop(['produceers', 'composers'], axis=1)
df_fillna_no_du
```

```python
df_fillna_no_du
```

```python
ko_dict = {'movieId':[], 'ko_genres':[], 'ko_plot':[]}
```

```python
for mid, g, p in tqdm(zip(df_fillna_no_du['movieId'][3203:], df_fillna_no_du['genres'][3203:], df_fillna_no_du['plot'][3203:])):
    try:
        g = trans_to_ko(g)
        g = g[2:-3]
    except TypeError:
        g = '장르 정보 없음'
    try:
        p = trans_to_ko(p)
        p = p[3:-3]
    except TypeError:
        p = '줄거리 정보 없음'
    ko_dict['movieId'].append(mid)
    ko_dict['ko_genres'].append(g)
    ko_dict['ko_plot'].append(p)   
```

```python
print(len(ko_dict['movieId']))
print(len(ko_dict['ko_genres']))
print(len(ko_dict['ko_plot']))
```

```python
ko_genres_plot_3229_df = pd.DataFrame(ko_dict)
ko_genres_plot_3229_df
```

```python
ko_genres_plot_3229_df.to_csv('/opt/ml/input/final_project/Data/Crawling/ko_genres_plot_3229_df.csv')
```
