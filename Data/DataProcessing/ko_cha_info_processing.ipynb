### 캐릭터 특징 크롤링 데이터 정제

```python
import pandas as pd
ko_cha = pd.read_pickle("/opt/ml/input/data/ko_character_info_df.pickle")
ko_cha.head()
```

```python
cha_df = pd.read_csv("/opt/ml/input/data/230129_Popular_movie_character_2867_cwj.csv")
cha_df.head()
```

```python
import copy
after_ko_cha = copy.deepcopy(ko_cha)
```

```python
after_ko_cha
```

```python
# 중국어인지 확인해주는 함수
def is_chinese(string):
    for char in string:
        if ord('\u4e00') <= ord(char) <= ord('\u9fff'): # Simplified Chinese
            return True
        if ord('\u3400') <= ord(char) <= ord('\u4DBF'): # Traditional Chinese
            return True
    return False
```

```python
def seperate(id, info):
    
    # info 컬럼 데이터에서 이름과 설명 분리
    name = info[info.find('이름:')+3:info.find(',', info.find('이름:')+3)].strip()
    description = info[info.find('설명:')+3:].strip()
    
    # 이름이 중국어일 경우 예외 처리
    if is_chinese(name): # 한글이나 영어가 아닐 경우 캐릭터의 원래 영어이름을 사용
        name = cha_df[cha_df['CharacterId'] == id]['Character'].values[0]
        
    print(name)
    print(description)
    return name, description
```

```python
after_ko_cha[['name', 'desc']] = after_ko_cha.apply(lambda row: seperate(row['characterId'], row['characterInfo']), axis=1, result_type='expand')
```

```python
after_ko_cha
```

```python
after_ko_cha = after_ko_cha[['characterId', 'name', 'desc']]
after_ko_cha.columns = ['CharacterId', 'name', 'desc']
after_ko_cha.head()
```
