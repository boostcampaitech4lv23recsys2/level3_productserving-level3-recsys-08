```python
import pandas as pd
import seaborn as sns
sim_df = pd.read_csv('enneagram_similarity.csv')
sim_df
```

## 가중치 변경
- 0과 1사이, 0.5부터라면 0과 1사이를 0~0.5사이로 바꿔야한다.

- 즉, 2로 나눠주고 0.5 더하기

```python
k=0.5
j=1/(1-k)
print(f"최소 유사도는 {k}")

processed_sim_df = sim_df.applymap(lambda x: k+(x/j))

processed_sim_df.style.background_gradient(
    cmap=sns.color_palette("Blues", as_cmap=True), low=0, high=0, axis=1, subset=df.columns
).format("{:.2f}%").set_properties(
    **{'font-size': '10pt', 'font-weight': 'bold', 'border-color': 'black', 'border-width': '0.3px'}
).set_caption('MBTI X ENNEGRAM')
```

```python
k=0.6
j=1/(1-k)
print(f"최소 유사도는 {k}, 최대는 {k+(1/j)}")

processed_sim_df = sim_df.applymap(lambda x: k+(x/j))

processed_sim_df.style.background_gradient(
    cmap=sns.color_palette("Blues", as_cmap=True), low=0, high=0, axis=1, subset=df.columns
).format("{:.2f}%").set_properties(
    **{'font-size': '10pt', 'font-weight': 'bold', 'border-color': 'black', 'border-width': '0.3px'}
).set_caption('MBTI X ENNEGRAM')
```
