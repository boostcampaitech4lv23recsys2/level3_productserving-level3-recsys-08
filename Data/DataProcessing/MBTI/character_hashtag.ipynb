## 캐릭터 별 MBTI 특징 태그 만들기

### 캐릭터별로 자신의 MBTI, Enneagram 특징 해시태그 컬럼 생성해주기

```python
import pandas as pd
import random
random.seed(8)
```

```python
character = pd.read_csv("/opt/ml/input/Data/DataProcessing/MBTI/Popular_movie_character_2867_with_MBTI2.csv", index_col=0)
character.head()
```

```python
ennea = pd.read_csv("/opt/ml/input/Data/DataProcessing/MBTI/enneagram_point.csv", index_col=0)
ennea
```

```python
mbti = pd.read_csv("/opt/ml/input/Data/DataProcessing/MBTI/mbti_point.csv", index_col=0)
mbti
```

```python
mbti['point_list'] = mbti['point'].apply(lambda x: list(x.split()))
```

```python
mbti.head()
```

enneagram 특징 2개와 mbti 특징 1개로 해시태그 만드는 함수

단, 랜덤으로 뽑은 mbti 특징이 enneagram 특징에 포함된다면 다시 랜덤 추출

```python
def make_hashtag(MBTI, Enneagram):
    ennea_cha = ennea[ennea['enneagram'] == Enneagram]['point'].values[0]
    tmp = random.choice(mbti[mbti['mbti'] == MBTI]['point_list'].values[0])
    while tmp in ennea_cha:
        tmp = random.choice(mbti[mbti['mbti'] == MBTI]['point_list'].values[0])
    mbti_cha = tmp
    return ennea_cha + ' ' + mbti_cha
```

기존 데이터프레임에 해시태그 컬럼 생성

```python
character['hashtag'] = character.apply(lambda x: make_hashtag(x.MBTI, x.Enneagram), axis='columns')
```

```python
character.head()
```

최종 데이터프레임에 덧붙여서 데이터프레임으로 저장

```python
# 최종 데이터프레임에 덧붙여서 데이터프레임으로 저장
character.to_csv("Popular_movie_character_2867_with_MBTI2_hashtag.csv")
```

{'캐릭터 ID' : '해시태그'} 딕셔너리만 pickle 파일로 저장

```python
character_hashtag_dic = {}
for id, tag in zip(character['CharacterId'].values, character['hashtag'].values):
    character_hashtag_dic[id] = tag
```

```python
character_hashtag_dic
```

```python
# {'캐릭터 ID' : '해시태그'} 딕셔너리만 pickle 파일로 저장
import pickle

with open('hashtag.pkl','wb') as f:
    pickle.dump(character_hashtag_dic,f)
```
