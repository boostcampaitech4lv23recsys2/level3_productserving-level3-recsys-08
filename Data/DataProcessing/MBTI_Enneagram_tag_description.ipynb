## MBTI와 Enneagram 조합의 성격 설명과 태그 pickle 파일 만들기

## 결과 파일 이미지


```python
import pandas as pd
```

```python
# 해당 불러오는 파일 위치 : https://docs.google.com/spreadsheets/d/1dPZ_MZ3O5RctlTNR3Fe5ZeaLwBUJL7Yf5BqlSFqVYGM/edit#gid=0
info_df = pd.read_csv('MBTIx애니어그램_성격과태그.csv')
```

```python
info_df = info_df[['MBTI','Enneagram','성격','태그']]
```

```python
info_df
```

```python
info_df['MBTI'].unique()
```

```python
info_df['Enneagram'].unique()
```

```python
info_df['MBTI_Enneagram'] = info_df['MBTI'] + ' ' + info_df['Enneagram']
```

```python
info_df['MBTI_Enneagram'].nunique() #288 = 16 * 18
```

```python
info_df['description'] = info_df['성격'].apply(lambda x: str(x).split("^^"))
```

```python
info_df[info_df['MBTI_Enneagram'] == 'ISTJ 1w9']['description'].values[0]
```

```python
info_df['tag'] = info_df['태그'].apply(lambda x: str(x).split("^^"))
```

```python
result_df = info_df[['MBTI_Enneagram','description','tag']]
```

```python
result_df
```

```python
result_df.to_csv('MBTI_Enneagram_personality_tag.csv',index=False)
```

```python
result_df.to_pickle('MBTI_Enneagram_personality_tag.pickle')
```
