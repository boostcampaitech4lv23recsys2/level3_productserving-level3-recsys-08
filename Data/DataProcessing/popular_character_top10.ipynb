## 인덱스 페이지 - 인기있는 캐릭터 Top 10
### contents_year, npop, Votes 를 이용해 스코어링 적용 후 정렬해서 Top 10 추출

```python
import pandas as pd
```

```python
data_movie = pd.read_csv("/opt/ml/input/data/230129_Popular_movie_1192_cwj.csv")
data_movie = data_movie[['movieId', 'Contents', 'contents_year', 'npop']]
data_movie.head()
```

```python
data_character = pd.read_csv("/opt/ml/input/Data/DataProcessing/MBTI/Popular_movie_character_2867_with_MBTI2_hashtag.csv", index_col=0)
data_character = data_character[['CharacterId', 'Character', 'movieId', 'Votes']]
data_character.head()
```

```python
data = data_movie.merge(data_character, on='movieId', how='inner')
data
```

연도  >= 2018

```python
data = data[data['contents_year'] >= 2018]
```

인기도 >= 20000

```python
data = data[data['npop'] >= 20000]
```

```python
data.sort_values('npop', ascending=False)
```

'Votes' featuer 스케일링을 위한 컬럼 전처리

```python
def make_votes(vote):
    vote = vote[:vote.index('/')]
    if vote[-1] == 'k':
        vote = int(vote[:-1].replace('.','')+'000')
    else:
        vote = int(vote)
    return vote
```

```python
data['clean_votes'] = data['Votes'].apply(lambda x: make_votes(x))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(data['clean_votes'].values.reshape(-1, 1))
scaled = scaler.transform(data['clean_votes'].values.reshape(-1, 1))
scaled
```

```python
data['scaled_votes'] = scaled
```

```python
data
```

```python
data.sort_values('scaled_votes', ascending=False)
```
