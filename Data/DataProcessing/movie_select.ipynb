#### 장르 원핫인코딩 컬럼 추가

```python
def delete_s(sen):
    sen = sen.replace(" ","")
    return sen
```

```python
def delete(genre_sentence):
    try:
        genre_sentence = genre_sentence.replace("'","")
        genre_sentence = genre_sentence[genre_sentence.index("[")+1:genre_sentence.index("]")]
        genre_sentence = list(map(delete_s, genre_sentence.split(",")))
        return genre_sentence
    except:
        return ["no_genre"]
```

```python
data['genres_list'] = data['genres'].apply(lambda x:delete(x))
```

```python
data.head()
```

```python
genre = []
for genres in list(data['genres_list']):
    genre.extend(genres)
genre = list(set(genre))
genre
```

```python
len(genre)
```

```python
values = []
for x in range(len(genre)):
    values.append(x)
print(values)
```

```python
import numpy as np
encoding = np.eye(len(values))[values]
encoding
```

```python
encoding.shape
```

```python
genre_dic = dict()
for idx, g in enumerate(genre):
    genre_dic[g] = idx
genre_dic
```

```python
def get_encoding(genre):
    g_encoding = np.zeros(28)
    for g in genre:
        g_encoding += encoding[genre_dic[g]]
    return g_encoding
```

```python
data['genre_encode'] = data['genres_list'].apply(lambda x: get_encoding(x))
```

```python
data.head()
```

```python
data['movieId']
```

```python
data['genre_encode'][0]
```

```python
encode_df = pd.merge(data['movieId'], pd.DataFrame(list(data['genre_encode']), index=data['movieId']), on='movieId',how='inner')
```

```python
encode_df
```

```python
X = encode_df.iloc[:,1:]
X
```

```python
# kmeans
from sklearn import cluster 
from sklearn import preprocessing
kmeans = cluster.KMeans(n_clusters=3, random_state=2023)
```

```python
X = preprocessing.StandardScaler().fit_transform(X)
kmeans.fit(X)
cluster_label = kmeans.labels_ 
```

```python
encode_df['cluster1'] = cluster_label
```

```python
# DBSCAN 모델 설정
# 한 데이터 포인트에서 eps 거리 안에 데이터가 min_samples 개수만큼 포함되어 있으면 클러스터로 포함
# eps 핵심 포인트를 중심으로 측정되는 유클리디언 거리값
# min_samples : 핵심 포인트를 중심점으로 간주하는 주변 지역의 표본 수
# X = preprocessing.StandardScaler().fit_transform(X)
dbm = cluster.DBSCAN(min_samples=10)
```

```python
dbm.fit(X)
cluster_label2 = dbm.labels_ # dbm.labels_ 중 -1은 noise
```

```python
encode_df['cluster2'] = cluster_label2
```

```python
encode_df
```

```python
encode_df['cluster1'].value_counts()
```

```python
encode_df['cluster2'].value_counts()
```

```python
data = pd.merge(data, encode_df[['movieId','cluster1']], on='movieId', how='inner')
```
