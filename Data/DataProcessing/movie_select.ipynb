### 연도 2000년 이상인 영화

```python
year = ml['title'].apply(lambda x: x.strip()[-5:-1])
year_li = []
for y in year:
    try:
        year_li.append(int(y))
    except:
        year_li.append(0)
```

```python
len(year_li)
```

```python
ml['year'] = year_li
```

```python
ml.head()
```

```python
ml_2000 = ml[ml['year'] >= 2000]
```

#### 장르 고려

```python
genre_li = []
def get_new_genre(genre):
    if genre == "(no genres listed)":
        genre = []
    else: 
        genre = list(genre.split("|"))
        genre_li.extend(genre)
    return genre
```

```python
ml_select['new_genre'] = ml_select['genres'].apply(lambda x : get_new_genre(x))
ml_select['new_genre']
```
